### Euclidean Distance of Toyota Valenzuela to City Centers in NCR

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from geopy.distance import great_circle
import contextily as ctx
```

```python
# Toyota Valenzuela (latitude, longitude)
toyota_valenzuela = (14.7519, 120.9789)

# Metro Manila city centers (approximate)
cities = {
    "Caloocan": (14.7568, 120.9822),
    "Las Piñas": (14.4549, 120.9800),
    "Makati": (14.5547, 121.0244),
    "Malabon": (14.6760, 120.9510),
    "Mandaluyong": (14.5794, 121.0359),
    "Manila": (14.5995, 120.9842),
    "Marikina": (14.6508, 121.1029),
    "Muntinlupa": (14.4140, 121.0387),
    "Navotas": (14.6555, 120.9413),
    "Parañaque": (14.4978, 121.0651),
    "Pasay": (14.5352, 121.0236),
    "Pasig": (14.5764, 121.0851),
    "Quezon City": (14.6760, 121.0437),
    "San Juan": (14.6006, 121.0358),
    "Taguig": (14.5176, 121.0509),
    "Valenzuela": (14.7065, 120.9818)  # city center of Valenzuela itself
}
```

```python
data = []
for city, coords in cities.items():
    dist_km = great_circle(toyota_valenzuela, coords).kilometers
    data.append({
        "city": city,
        "lat": coords[0],
        "lon": coords[1],
        "distance_km": dist_km
    })
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 10))

# Scatter all city centers
sns.scatterplot(
    x="lon",
    y="lat",
    size="distance_km",
    sizes=(50, 300),
    hue="distance_km",
    palette="viridis_r",
    data=df,
    legend="brief"
)

# Mark Toyota Valenzuela
plt.scatter(
    toyota_valenzuela[1],
    toyota_valenzuela[0],
    color="red",
    marker="X",
    s=200,
    label="Toyota Valenzuela"
)

# Annotate distances
for idx, row in df.iterrows():
    plt.text(
        row["lon"] + 0.005,
        row["lat"] + 0.005,
        f"{row['city']} ({row['distance_km']:.1f} km)",
        fontsize=8
    )

# Basemap (OpenStreetMap)
ctx.add_basemap(
    plt.gca(),
    crs="EPSG:4326",
    source=ctx.providers.OpenStreetMap.Mapnik
)
```
